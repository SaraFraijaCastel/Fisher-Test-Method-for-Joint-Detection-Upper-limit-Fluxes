```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Combine the significance values
A = data_1['Significance'].to_list() + data_2['Significance'].to_list()

# Define bins
bins = np.log(np.arange(0, 6, 0.5))

# Compute mean and standard deviation
mu, sigma = np.mean(A), np.std(A)

# Create histogram
plt.hist(A, bins=bins, density=True, alpha=0.6, color='b', label="Histogram")

# Generate values for the normal distribution fit
x = np.linspace(min(A), max(A), 100)
pdf = norm.pdf(x, mu, sigma)

# Plot the fitted normal distribution
plt.plot(x, pdf, 'r-', linewidth=2, label=f'Normal Fit\n$\\mu={mu:.2f}$, $\\sigma={sigma:.2f}$')

# Labels and legend
plt.xlabel('Significance')
plt.ylabel('Density')
plt.legend()
plt.title('Histogram with Normal Fit')

# Show plot
plt.show()
```
