```python
import pandas as pd
```

```python
GRBsINFO_1='/lustre/hawcz01/scratch/userspace/jorgeamontes/GRB_KN/data/ULs/config/Coordinates/PSF_0.3/alfa=2.07/Coordinates_with_Max_Sig_1_00.3.csv'
GRBsINFO_2='/lustre/hawcz01/scratch/userspace/jorgeamontes/GRB_KN/data/ULs/config/Coordinates/PSF_0.3/alfa=2.07/Coordinates_with_Max_Sig_2_00.3.csv'
Energy_ranges='/lustre/hawcz01/scratch/userspace/jorgeamontes/GRB_KN/config/Energy_ranges.csv'
GPS_times='/lustre/hawcz01/scratch/userspace/jorgeamontes/GRB_KN/config/newgrblist.csv'
Table='/lustre/hawcz01/scratch/userspace/jorgeamontes/GRB_KN/data/ULs/Latex/Tables_NT/Table_1.txt'
```

```python
len(GPS_times)
```

```python

T1=pd.read_csv(GRBsINFO_1)
T2=pd.read_csv(GRBsINFO_2)
E=pd.read_csv(Energy_ranges)
T1=T1[['Name','Ra','Dec','z','Error_Radius']]
T2=T2[['Name','Ra','Dec','z','Error_Radius']]
Times=pd.read_csv(GPS_times)
df_merged=pd.merge(left=T1,right=T2,how='outer')
df_merged=df_merged.fillna('-')
df_merged_energy=pd.merge(left=df_merged,right=E,how='inner')
caption='Final Selection of GRBs, including their coordinates in degrees, their GBM error radius, and redshifts (if known). Redshifts with  an asterisk (*) are pseudo-redshifts from (\cite{2018ApJ...852....1Z}) and those marked with two asterisks (**) are from (\cite{2023ApJ...943..126D}). }\label{tab:GRBs} '
Times[r'$T_{HAWC} (h)$']=(Times['First Transit Start MJD']-Times['TRIG MJD'])*24
Times.loc[Times[r'$T_{HAWC} (h)$']<0,r'$T_{HAWC} (h)$']=0
T=Times[[r'$T_{HAWC} (h)$','Name']]
df_final=pd.merge(left=df_merged_energy,right=T,left_on='Name',right_on='Name')
df_final.rename(columns={'Error_Radius':'$\Delta$ r $(^{\circ})$'},inplace=True)
df_final.rename(columns={'Ra':'Ra $(^{\circ})$'},inplace=True)
df_final.rename(columns={'Dec':'Dec $(^{\circ})$'},inplace=True)
df_final.rename(columns={'minE':r'E$_{min}$ (TeV)'},inplace=True)
df_final.rename(columns={'maxE':r'E$_{max}$ (TeV)'},inplace=True)
df_final.loc[(df_final['z']==0.009)&(df_final['Name']!='GRB170817529'),'z']='-'
df_final=df_final.round(2)
df_final.sort_values(by='Name',inplace=True)
df_final
```

```python
C1_DIR='/lustre/hawcz01/scratch/userspace/smfraijac/Catalogs/pseudoredshifts/Catalog1.txt'
C2_DIR='/lustre/hawcz01/scratch/userspace/smfraijac/Catalogs/pseudoredshifts/Catalog2.txt'
ICECUBE='/lustre/hawcz01/scratch/userspace/smfraijac/Catalogs/GRBs_ICECUBE/ICECUBE_catalog.csv'
GRBs_selection='/lustre/hawcz01/scratch/userspace/jorgeamontes/GRB_KN/data/ULs/Latex/Tables_NT/Table.csv'
column_names_1= ['name', 'z_est_1', 'T_90', 'log F_X', 'S (15-150 keV)',
    'log (T_a/10^3)', 'log (L_X/10^47)', 'log (E_gamma,iso/10^53)',
    'alpha_1', 'alpha_2'
]
column_names_2= [
    "GRB", "T90", "z_est_2", "Flux", "e_Flux", "Ep", "e_Ep",
    "alpha", "beta", "Lum", "e_Lum"
]
GRBs=df_final
ICE=pd.read_csv(ICECUBE)
C1=pd.read_csv (C1_DIR,sep='\t',skiprows=[2],header=1)
C1['source']='\cite{Pseudo1}'
C2=pd.read_csv(C2_DIR,sep='\s+',skiprows=13,names=column_names_2)
C2['source']='\cite{Pseudo2}'
Final=pd.merge(left=C1,right=ICE,on='name',how='outer')
Final2=pd.merge(left=Final,right=C2,left_on='Fermi_Name',right_on='GRB',how='outer')
F=Final2[['name','Fermi_Name','z_est_1','z_est_2']]
final=pd.merge(left=GRBs,right=F,left_on='Name',right_on='Fermi_Name',how='left')
final.fillna('-',inplace=True)
final.loc[final['z']=='-','z']=final[final['z']=='-']['z_est_2'].apply(lambda x:str(x)+'*' if x!='-' else '-')
final.loc[final['z']=='-','z']=final[final['z']=='-']['z_est_1'].apply(lambda x:str(x)+'**' if x!='-' else '-')
final.drop(columns={'z_est_1','z_est_2','name', 'Fermi_Name'},inplace=True)
```
